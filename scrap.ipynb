```python
"""
# Filename: scrap.ipynb.py
"""

## Run selenium and chrome driver to scrape data from Wikipedia.com
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
```

```python
## Setup chrome options
chrome_options = Options()
chrome_options.add_argument("--headless") # Ensure GUI is off
chrome_options.add_argument("--no-sandbox")

# Set path to chromedriver as per your configuration
webdriver_service = Service("/home/standm/dev/presidentials/chromedriver")

# Choose Chrome Browser
browser = webdriver.Chrome(service=webdriver_service, options=chrome_options)


```

```python
# Get page
browser.get("https://fr.wikipedia.org/wiki/Candidats_%C3%A0_l%27%C3%A9lection_pr%C3%A9sidentielle_fran%C3%A7aise_de_2022")

# Extract description from page and print
description = browser.find_element(By.CLASS_NAME, "wikitable")
df = pd.read_html(description.get_attribute("outerHTML"))


#Wait for 10 seconds
time.sleep(7)
browser.quit()
```

```python
df_candidates = pd.DataFrame(data= df[0], index =None)
```
